```python
annGenes = trinotateResults.loc[(trinotateResults["sprot_Top_BLASTX_hit"] != ".") | (trinotateResults["sprot_Top_BLASTP_hit"] != ".") | (trinotateResults["Pfam"] != ".")]
```

```python
annTranscripts = set(annGenes["transcript_id"].tolist()) #non-duplicate transcript ids of annotated sequences
```

```python
outfilePath = "/home/nnp9/project/podocoryna/results/trinotate/transcripts_with_annotation_nodupes.txt"
with open (outfilePath, "w") as outfile:
    for item in annTranscripts:
        outfile.write(item + "\n")
```

```python
# List the transcripts ids that blast on the reverse strand
searchStr = ".+-.+[-]"
negGenes = []
for value in annGenes["prot_coords"].values:
    if re.match(searchStr,value):
        transcriptID = annGenes.loc[annGenes["prot_coords"] == value, 'transcript_id'].tolist()
        negGenes.append(transcriptID)
```

```python
# Convert a list of lists to a flat list
negGenes_flat = []
for sublist in negGenes:
    for item in sublist:
        negGenes_flat.append(item)
```

```python
# Drop duplicates and use original list name
# negGenes = list(set(negGenes_flat))
len(negGenes)
```

# Some of these transcripts have both positive and negative hits. Let's select  the ones that only have negative hits

```python
# Select rows if value for transcript id is in negGenes list
negSubsetAnnGenes = annGenes[annGenes['transcript_id'].isin(negGenes)]
```

```python
negSubsetAnnGenes.head(30)
```

```python
# Create a dictionary with key, values (transcript id: [prot coords]) that keeps multiple values for each key
multivalue_dict = defaultdict(list)

for idx,row in negSubsetAnnGenes.iterrows():
    multivalue_dict[row['transcript_id']].append(row['prot_coords'])
```

```python
multivalue_dict
```

```python
# Drop key if one of the values has a + sign on it


# for keys that have more than one value, see if one of the values is positive, if it is, delete the key
multipositives = []
negonly = []
searchStr = ".+-.+[+]"
r = re.compile(searchStr)
for name, coordinates in multivalue_dict.items():
    if len(coordinates) > 1:
        if len(list(filter(r.match, coordinates))) > 0:
            multipositives.append(name)
            print(name, coordinates)
#     else:
#         multiprot.append(name)
```
